```python
from matplotlib import pyplot as plt
from plot import *
import numpy as np
print(np.array(img).shape, np.array(hm).shape)
print(np.array(so_x).shape)
def overlay(img, over, alpha=0.5):
    out = img.copy()
    if img.max() > 1.:
        out = out / 255.
    out *= 1-alpha
    if len(over.shape)==2:
        out += alpha*over[:,:,np.newaxis]
    else:
        out += alpha*over    
    return out

idx = 1
plt.figure()
plt.imshow(overlay(img[idx], np.max(hm[idx], axis=-1), alpha=0.5))
```
