```python
import config
import random
import tensorflow as tf
import numpy as np
import cv2 as cv
from pycocotools.coco import COCO
import skimage.io as io
import matplotlib.pyplot as plt
import pylab

ann_file = '../KerasPersonLab/ANNO_FILE/person_keypoints_train2017.json'
img_dir = '../KerasPersonLab/IMG_DIR/train2017/'
coco = COCO(ann_file)
cat_ids = coco.getCatIds(catNms=['person'])
img_ids = coco.getImgIds(catIds=cat_ids)
img_infos = coco.loadImgs(img_ids)
```

```python
idx = 333
img_info = img_infos[idx]
img_id = img_info['id']
img_path = img_dir + '/' + img_info['file_name']
ann_ids = coco.getAnnIds(imgIds=img_id, catIds=cat_ids)
anns = coco.loadAnns(ann_ids)

I = io.imread(img_info['coco_url'])
plt.imshow(I)
plt.axis('off')
```

```python
I = io.imread(img_info['coco_url'])
#I = cv.imread(img_path, cv.IMREAD_COLOR)
for a in anns:
    a['nkp'] = a['keypoints'][2::3].count(2)
    print(a['area'], a['nkp'],  a['num_keypoints'])
sel_anns = [ann for ann in anns if ann['area'] > 3000 and ann['nkp'] >= 5 and not ann['iscrowd']]
plt.rcParams["figure.figsize"] = (20, 10)
plt.imshow(I)
plt.axis('off')
coco.showAnns(sel_anns)
```
